```python
import pandas as pd
import lxml as lxml
import numpy as np
import html5lib as ht
from pandas.io.html import read_html
```

```python
pl1 = 'https://www.moneycontrol.com/financials/interglobeaviation/profit-lossVI/IA04/#IA04'
pl2 = 'https://www.moneycontrol.com/financials/interglobeaviation/profit-lossVI/IA04/2#IA04'
#pl3 = 'https://www.moneycontrol.com/financials/interglobeaviation/profit-lossVI/IA04/3#IA04'
```

```python
p1 = read_html(pl1,  attrs={"class":"mctable1"}, index_col=0, header=0)
p1 = pd.DataFrame(p1[0])
p2 = read_html(pl2,  attrs={"class":"mctable1"}, index_col=0, header=0)
p2 = pd.DataFrame(p2[0])
#p3 = read_html(pl3,  attrs={"class":"mctable1"}, index_col=0, header=0)
#p3 = pd.DataFrame(p3[0])
```

```python
index = p1.index.name
```

```python

p = p1.merge(p2, on = index)
#p = pt.merge(p3, on = index)
p.head()
```

```python
p= p.drop(p.columns[[5,-1]], axis = 1)
p = p.drop([p1.index[0]])
p = p.replace(np.nan, "")
p = p.sort_index(axis=1 ,ascending=True)
p = p.convert_objects(convert_numeric=True)
type(p1)
```

```python
corr = p.T.corr()
r =  corr.iat[1,1]
corr = corr.replace(to_replace = r, value = int(.001))
exp  = corr.idxmax(axis=1)
```
