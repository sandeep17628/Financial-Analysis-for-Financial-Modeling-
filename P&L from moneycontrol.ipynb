```python
import pandas as pd
import lxml as lxml
import numpy as np
import html5lib as ht
from pandas.io.html import read_html
```

```python
pl1 = 'https://www.moneycontrol.com/financials/cipla/profit-lossVI/C#C'
pl2 = 'https://www.moneycontrol.com/financials/cipla/profit-lossVI/C/2#C'
pl3 = 'https://www.moneycontrol.com/financials/cipla/profit-lossVI/C/3#C'
```

```python
p1 = read_html(pl1,  attrs={"class":"mctable1"}, index_col=0, header=0)
p1 = pd.DataFrame(p1[0])
p2 = read_html(pl2,  attrs={"class":"mctable1"}, index_col=0, header=0)
p2 = pd.DataFrame(p2[0])
p3 = read_html(pl3,  attrs={"class":"mctable1"}, index_col=0, header=0)
p3 = pd.DataFrame(p3[0])
pt = p1.merge(p2, on = "Profit & Loss account of Cipla (in Rs. Cr.)")
p = pt.merge(p3, on = "Profit & Loss account of Cipla (in Rs. Cr.)")
```

```python

p = p.drop(columns ="Unnamed: 6")
p = p.drop(columns ="Unnamed: 6_x")
p = p.drop(columns ="Unnamed: 6_y")
p = p.drop([p1.index[0]])
p = p.replace(np.nan, "")
p = p.sort_index(axis=1 ,ascending=True)
type(p1)
```
